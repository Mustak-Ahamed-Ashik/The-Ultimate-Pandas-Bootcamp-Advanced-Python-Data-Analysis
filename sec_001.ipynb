```python
# Q. Multiply all the numbers by 10

# Method-1
arr = []
for i in list_name:
  arr.append(i*10)

print(arr)

# Method-2
print(list(map(lambda i: i*10, list_name)))

# Method-3
print([i*10 for i in list_name])
```
